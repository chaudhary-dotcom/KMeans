```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_clustering.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['cgpa'], df['iq'])
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Students Clustering')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)

wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.show()
```

```python
X = df.iloc[:, :].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
X[y_means == 3]  # [3] is the cluster number 
X[y_means == 3, 1]  # It means we eXtract the value of cluster number 3 with 1 column (we have two column)
```

```python
plt.scatter(X[y_means == 0,0], X[y_means == 0,1], color='blue')
plt.scatter(X[y_means == 1,0], X[y_means == 1,1], color='red')
plt.scatter(X[y_means == 2, 0], X[y_means == 2,1], color ='green')
plt.scatter(X[y_means == 3,0], X[y_means == 3,1], color = 'yellow')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Students Clustering')
plt.show()
```

# K-Means on 3D Data

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5), (5,5,-5), (3.5, -2.5, 4), (-2.5, 2.5, -4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200, cluster_std=cluster_std, centers=centroids, n_features=3, random_state=1)
```

```python
X
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'browser'  # Force browser view

fig = px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2])
fig.show()
```
